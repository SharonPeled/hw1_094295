```python
import pandas as pd
import numpy as np 
import json
import ast
from collections import Counter
# pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
DATA_FOLDER = "/StudentData/"
```

```python
train = pd.read_csv(DATA_FOLDER + "train_reformated.csv")
train.head()
```

```python
def numeric_full_columns(df):
    # finds all the columns who are numeric and withour nulls - "ready for being features"
    numeric_columns = []
    for col in df.columns:
        if np.issubdtype(df[col].dtype, np.number) and df[col].isna().sum()==0:
            numeric_columns.append(col)
    return numeric_columns
            


def generate_feature_matrix(df_input):
    df = df_input.copy()
    y = df["revenue"]
    df.drop(columns=["revenue"],inplace=True) # so we won't use it as feature
    X = pd.DataFrame(index=df.index)
    X = X.join(df[numeric_full_columns(df)])
    return X, y
```

```python
X_train, y_train = generate_feature_matrix(train)
X_train.head()
```

```python
X_train.to_csv(DATA_FOLDER + "X_train.csv",index=False)
y_train.to_csv(DATA_FOLDER + "y_train.csv",index=False)
```
